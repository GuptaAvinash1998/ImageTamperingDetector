```python
# we want to have only JPG or JPEG files for our model and nothing else, go through each of the images and
# get only the JPG or JPEG files. We are only taking JPEG files because we want to deal with formats that are
# not lossless. Lossless means that when the file is compressed and decompressed, no imformation or quality of the
# data is lost.
# Having lossless images in this model will be difficult because one of that we are detecting image tampering
# is through compression differences, so if we have formats that do not loose any data, it will be hard to train
# the model so we ignore these formats
```

```python
# load a few samples of the authentic and tampered images
def createImageGrid(imagePath):
    
    #sample 4 images from the path
    imageList = random.sample(os.listdir(imagePath), 4)
    
    #create a 2x2 grid
    fig, axis = plt.subplots(2, 2, figsize=(10,6))
    
    #load each image in the axis and display the image
    for i, ax in enumerate(axis.flatten()):
        
        #load the image
        #This gets the full path the file
        im_path = os.path.join(imagePath, imageList[i])
        
        #read the image
        image = plt.imread(im_path)
        
        # display the image
        ax.imshow(image)
        
        # turn off axis labeling
        ax.axis('off')
        
    # create a tight layout for a compressed grid
    plt.tight_layout()
    
    plt.show()
```

```python
# preprocess the image data
# the preprocessing step creates ELA images so that we filter out the unwanted file formats
# ELA aka Error Level Analysis is a forensic technique that is used to analyze images by comparing
# them through different levels of compression.
def preProcessImages(imagesPath, files, resavePath):
    
    for file in files:
        imagePath = os.path.join(imagesPath, file)
        
        #we accept 3 lossy file extensions: jpg, jpeg, and png. Process images in this formar only
        if file.endswith('jpg') or file.endswith('jpeg') or file.endswith('png'):
            
            #open the image and convert it to RGB mode
            image = Image.open(imagePath).convert('RGB')
            
            #resave the image with the new mode and as a jpg
            # when we save an image with a quality of 90, that means we store the image
            # with high quality but moderate compression
            image.save('resavedImage.jpg', 'JPEG', quality=90)
            
            # reopen the resaved image
            resavedImage = Image.open('resavedImage.jpg')
            
            # Calculate the ELA by taking the difference between the originial and the resaved image
            elaImage = ImageChops.difference(image, resavedImage)
            
            # get the maximum pixel value
            maxPixelValue = max([val[1] for val in elaImage.getextrema()])
            
            # if the max value is 0, set it to 1 so that we do not divide by 0
            if(maxPixelValue <= 0):
                maxPixelValue = 1
                
            # using the max value, scale the pixels to the range of [0,255]
            scaleVal = 255.0/maxPixelValue
            
            #using the scaled value, enhance the image and save it
            elaImage = ImageEnhance.Brightness(elaImage).enhance(scaleVal)
            
            #save the image in the reserved file path as a jpeg
            elaImage.save(os.path.join(resavePath, os.path.basename(imagePath)), 'JPEG')
            
```

```python
# create the augmented images using the ImageDataGenerator
imageDataGenerator = ImageDataGenerator()

generatorBasePath = os.path.join(os.getcwd(), 'train')

# The way the image generator works, is it takes a directory as input and based on the names of the subdirectories that the images
# are in, it will automatically label the images with those names. In this case, since our training images are in the train/Authentic and
# train/Tampered, it will go through these images automatically label the images in the Authentic directory as Authentic and the images in
# the Tampered directory as Tampered. same goes with the validation dataset

# we are going to create 2 generators, for images of size 256x256
# images of size 256x256 are a good balance between data complexity and model efficency

# these generators will generate images that are of size 256x256
dataTrainGen256 = imageDataGenerator.flow_from_directory(os.path.join(os.getcwd(), 'train'), target_size=(256, 256), class_mode='binary')
dataValGen256 = imageDataGenerator.flow_from_directory(os.path.join(os.getcwd(), 'val'), target_size=(256, 256), class_mode='binary')
```

```python
# create the model for classification

# we are going to combine 2 models, the InceptionnetV3 and the VGG16 models
# to do this, we need to do the following

# 1: instantiate the models
# This creates the InceptionV3 network. The first argument says that we do not want the fully connected layer on top as the
# last layer in teh network. The second argument says that the model will take in a shape of 256, 256 as the width and height of the images
# and 3 for the number of channels the images has.
# The network will by default have the weights that were generated when training with the imagenet dataset
inceptionNet = InceptionV3(include_top=False, input_shape=(256,256,3))

# This creates the VGG16 network. The first argument says that we do not want the fully connected layer on top as the
# last layer in teh network. The second argument says that the model will take in a shape of 256, 256 as the width and height of the images
# and 3 for the number of channels the images has.
# The network will by default have the weights that were generated when training with the imagenet dataset
vggNet = VGG16(include_top=False, input_shape=(256,256,3))

# 2: freeze the trainable layers
# After instantiating the model, we want to keep the weight of the model from imagent and not retrain it. If we do not freeze the trainable
# parameters, all the values will be reset and we will loose the default weights

# for each layer in the models, set the trainable layer to false
# the InceptionNet has 311 layers and the VGG net has 19 layers

for layer in inceptionNet.layers:
    layer.trainable = False
    
for layer in vggNet.layers:
    layer.trainable = False
```

```python
print(vggNet.summary())
```

```python
# 3: Create the fusion model

# create an instance of the input layer
inputLayer = tf.keras.Input(shape=(256,256,3))
```

```python
# pass the input layer through the InceptionNet layers. This results in extrating the features of the input images
googleNetFeatures = inceptionNet(inputLayer)
```

```python
# do the same to the vgg model
vggFeatures = vggNet(inputLayer)
```

```python
# 4: Reshape the inception net to match the dimensions of the vgg model
googleNetFeatures = layers.GlobalAveragePooling2D()(googleNetFeatures)
print(googleNetFeatures)
googleNetFeatures = layers.Dense(1024, activation='relu')(googleNetFeatures)
print(googleNetFeatures)
googleNetFeatures = tf.expand_dims(googleNetFeatures, axis=1)
print(googleNetFeatures)
googleNetFeatures = tf.expand_dims(googleNetFeatures, axis=1)
print(googleNetFeatures)
googleNetFeatures = layers.UpSampling2D(size=(8,8))(googleNetFeatures)
print(googleNetFeatures)
```

```python
# 5: concatenate the 2 models
fusedFeatures = layers.Concatenate()([googleNetFeatures, vggFeatures])
print(fusedFeatures)
```

```python
# 6: add additional layers for fusion and prediction
fusedFeatures = layers.GlobalAveragePooling2D()(fusedFeatures)
print(fusedFeatures)
fusedFeatures = layers.Dense(1024, activation='relu')(fusedFeatures)
print(fusedFeatures)
```

```python
predictionLayer = layers.Dense(1, activation='sigmoid')(fusedFeatures)
print(predictionLayer)
```

```python
# 7: create the fusion model
fusedModel = models.Model(inputs=inputLayer, outputs=predictionLayer)
```

```python
# 8: compile the model
fusedModel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print(fusedModel.summary())
```

```python
#train the model from the generated images
# the first parameter is the generated data
# the second parameter is the number of batches to process from per epoch. Here we are taking the batch size to be 8
# the third parameter is the number of epochs
# the fourth parameter is the validation dataset
history = fusedModel.fit(dataTrainGen256,  steps_per_epoch=len(dataTrainGen256) / 8, epochs=100, validation_data = dataValGen256)
```

```python
def plotAccuracy():

    #get the training accuracy and validation accuracies
    trainAccuracy = history.history['accuracy']
    valAccuracy = history.history['val_accuracy']
    epochs = range(1, len(trainAccuracy)+1)

    # create the plot, draw the train accuracy in blue and the val accuracy in red
    plt.plot(epochs, trainAccuracy, 'b', label='Training Accuracy')
    plt.plot(epochs, valAccuracy, 'r', label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracies')
    plt.legend()

    plt.show()
```

```python
plotAccuracy()
```

```python
def plotLoss():

    #get the training accuracy and validation loss
    trainLoss = history.history['loss']
    valLoss = history.history['val_loss']
    epochs = range(1, len(trainLoss)+1)

    # create the plot, draw the train accuracy in blue and the val accuracy in red
    plt.plot(epochs, trainLoss, 'b', label='Training Loss')
    plt.plot(epochs, valLoss, 'r', label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()
```

```python
plotLoss()
```

```python
def prepareForPrediction(sourcePathFolder, images, destinationPathFolder):

    #first pre process the image(s)
    preProcessImages(sourcePathFolder, images, destinationPathFolder)

    #using pillow, open the image, resize it to 256x256 because that is what the model expects
    #then add another dimension because the input dimension for the model is (None, 256, 256, 3)
    img = Image.open(os.path.join(destinationPathFolder, images[0]))
    
    # Resize the image to the input size expected by the model
    img = img.resize((256, 256))  # Assuming InceptionV3 expects input size of 299x299

    # Convert the image to a numpy array
    imgArray = np.asarray(img)

    # Reshape the array to match the expected input shape of the model
    #imgArray = np.expand_dims(img_array, axis=0)
    imgArray = imgArray[np.newaxis, :]

    return imgArray
```
